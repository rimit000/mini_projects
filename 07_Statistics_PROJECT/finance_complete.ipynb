```python
# 10년치 주가 및 배당금 데이터 가져오기
data = nvda.history(period="10y")
dividends = data["Dividends"]
close_prices = data["Close"]
```

```python
# 데이터 정리
df = pd.DataFrame({"Close Price": close_prices, "Dividends": dividends})

df
```

```python
# 상관관계 분석
correlation = df.corr()
print("배당금과 주가의 상관계수:\n", correlation)
```

```python
# 시각화
fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel("Date")
ax1.set_ylabel("Close Price (USD)", color="blue")
ax1.plot(df.index, df["Close Price"], color="blue", label="Close Price")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
ax2.set_ylabel("Dividends (USD)", color="red")
ax2.bar(df.index, df["Dividends"], color="red", alpha=0.6, label="Dividends")
ax2.tick_params(axis="y", labelcolor="red")

plt.title("엔비디아(NVDA) 배당금과 주가 비교")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. NaN 값 제거
df = df.dropna()  

# 2. 배당금 0이 아닌 경우만 사용
df = df[df["Dividends"] > 0]  

# 3. 로그 변환 (필요한 경우)
df["Log_Dividends"] = np.log1p(df["Dividends"])  # log(1 + x) 변환

# 4. 독립 변수 (X)와 종속 변수 (y) 설정
X = df["Log_Dividends"].values.reshape(-1, 1)  # 배당금 로그 변환
y = df["Close Price"].values  # 종가

# 5. 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# 6. 예측값 계산
y_pred = model.predict(X)

# 7. 결과 시각화
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="실제 주가", color="blue", alpha=0.5)
plt.plot(X, y_pred, label="회귀선", color="red", linewidth=2)
plt.xlabel("배당금 (Log 변환)")
plt.ylabel("주가 (Close Price)")
plt.title("배당금과 주가의 관계 (선형 회귀)")
plt.legend()
plt.show()
```
